```python
df_uber_pickups_april = pd.read_csv('datasets/required/uber-raw-data-apr14.csv')
df_uber_pickups_may = pd.read_csv('datasets/required/uber-raw-data-may14.csv')
df_uber_pickups_june = pd.read_csv('datasets/required/uber-raw-data-jun14.csv')
df_uber_pickups_july = pd.read_csv('datasets/required/uber-raw-data-jul14.csv')
df_uber_pickups_august = pd.read_csv('datasets/required/uber-raw-data-aug14.csv')
df_uber_pickups_sept = pd.read_csv('datasets/required/uber-raw-data-sep14.csv')

print("Done")
```
